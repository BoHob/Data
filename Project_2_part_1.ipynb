The following is the correlation between the price of homes and what year they were sold in. In the first code block you will see multiple variables, although I chose to share the price and year. 

```python
import pandas as pd 
import requests
url = 'https://raw.githubusercontent.com/BoHob/House-Prices/main/train.csv'
df = pd.read_csv(url)
df.T
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Below is the coorelation between the variables above, I looked at over 50+ variables in the pearson coorelation affect.

```python
df['SalePrice'].plot(kind='hist',figsize=(10, 5))
plt.title('Sales Price 2008-Present')
plt.ylabel('Number of Homes')
plt.xlabel('Price Range')
```

The graph above explains the number of homes sold over the years in 2008 to present, y-axis being number of homes x-axis being the price range.
