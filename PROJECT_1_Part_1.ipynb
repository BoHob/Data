```python
def Stock(ticker=True):
  url = 'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY_ADJUSTED&symbol='+ticker+'&apikey=SUDOM26XKQMJJJQB'
  r = requests.get(url)
  data= r.json()
  stock=pd.DataFrame.from_dict(data["Monthly Adjusted Time Series"], orient="index")
  stock.reset_index(inplace=True)
  stock.rename(columns={"1. open":"open","2. high":"high","3. low":"Low","4. close":"close","5. adjusted close":"close", "6. volume":"volume",
                        "7. dividend amount":"dividend","index":"date"},inplace=True)
  stock["ticker"]=ticker
  return stock
```

```python
Stock(ticker='A')
```

```python
import matplotlib.pyplot as plt

stocks=Stock(ticker="A")

stocks['date']=pd.to_datetime(stocks['date'])
stocks.dtypes

column_names=stocks.columns
for i in column_names[1:8]:
  stocks[i]= stocks[i].astype('float')

plt.plot(stocks['date'],stocks['high'])
```
